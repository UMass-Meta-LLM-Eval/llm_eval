```python
import json
from matplotlib import pyplot as plt
```

```python
with open('result_scores.json', 'r') as f:
    results = json.load(f)
```

```python
results
```

```python
result_vals = []
for v in results.values():
    if 'acc,none' in v:
        result_vals.append(v['acc,none'])
print(len(result_vals))
```

```python
fig, ax = plt.subplots()
ax.hist(result_vals)
fig.show()
```
